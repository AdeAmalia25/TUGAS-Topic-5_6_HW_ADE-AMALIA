# 4) Create a new ‘bmi’ column calculated based on the following formula
# BMI (kg/m^2) = weight / (height/100)^2

```python
# BMI (kg/m^2) = weight / (height/100)**2
```

```python
def bmi(weight, height):
    return weight / (height/100)**2
```

```python
df_dropbn["bmi"] = df_dropbn.apply(lambda x: bmi(x["weight"], x["height"]), axis = 1)
```

# 5) Create a new ‘bmi_class’ column following the definition in the table below.

```python
# BMI                                                                  Class

# < 18.5                                                       Underweight

# 18.5 <= BMI < 25                                             Normal weight

# 25 <= BMI < 30                                               Overweight

# 30 <= BMI < 35                                               Obesity class 1

# 35 <= BMI < 40                                               Obesity class 2

# >= 40                                                        Obesity class 3
```

```python
def bmi_class(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif bmi >= 18.5 and bmi < 25.0 :
        return "Normal Weight"
    elif bmi >= 25.0 and bmi < 30.0:
        return "Overweight"
    elif bmi >= 25.0 and bmi < 35.0:
        return "Obesity class 1"
    elif bmi >= 25.0 and bmi < 40.0:
        return "Obesity class 2"
    elif bmi >= 40.0:
        return "Obesity class 3"
    else:
        return np.nan
```

```python
df_dropbn["bmi class"] = df_dropbn.apply(lambda x: bmi_class(x["bmi"]), axis = 1)
```
